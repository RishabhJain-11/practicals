### Import python libraries

### Reading the Facebook metrics dataset





### Remove Nan values

```python
data = data.dropna()
data
```









## Create data subsets

### Create data subset for first 8 columns

```python
col1 = data.iloc[:,:8]
col1
```

```python
col1.shape
```

### Create data subset for remaining columns

```python
col2 = data.iloc[:,7:14]
col2
```

```python
col2.shape
```

### Create data subset for first 200 rows

```python
subset1 = data.iloc[:200]
subset1
```

```python
subset1.shape
```

### Create data subset for remaining rows

```python
subset2 = data.iloc[200:]
subset2
```

```python
subset2.shape
```

## Merge Data

### Merge the subset of first 8 columns with subset of remaining columns

```python
merged_data = pd.merge(col1, col2)
merged_data
```

```python
merged_data.shape
```

### Merge subset of first 200 rows with subset of remaining rows

```python
merged_data2 = pd.concat([subset1,subset2])
merged_data2
```

```python
merged_data2.shape
```
