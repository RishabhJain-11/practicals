### Import python libraries

```python
import pandas as pd
import numpy as np
```

### Reading the AirQuality Dataset

```python
df= pd.read_csv("airquality.csv")
df
```

```python
df.isnull().sum()
```

## Data Cleaning

### Fill the NaN values with mean of respective column

```python
df.columns
```

```python
df['Ozone'].fillna(df['Ozone'].mean(), inplace=True)
df['Solar.R'].fillna(df['Solar.R'].mean(), inplace=True)
df['Wind'].fillna(df['Wind'].mean(), inplace=True)
df["Humidity"].fillna(df["Humidity"].mode(), inplace = True)
df = df.drop(['Unnamed: 0'], axis = 1)
df
```

## Error Correcting

### Look for Missing Values

```python
df.isnull().sum()
```

```python
df.describe()
```

## Data Transformation

### Using Label Encoding for Encoding column values

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
df["Humidity"]= le.fit_transform(df["Humidity"])
df
```

## Data Model Building

### Splitting x and y data

```python
x= df.iloc[:, [0,3]].values
y= df["Humidity"]
```

### Splitting the Training and Testing Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25)
```

### Importing Linear Regression for prediction

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(x_train, y_train)
```

```python
prediction= model.predict(x_test)
prediction
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction)) 
```
