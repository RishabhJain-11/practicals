```python
import pandas as pd
import numpy as np
df = pd.read_csv("airquality (1).csv")
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
df.isnull().sum()
```

```python
df['Ozone'].fillna(df['Ozone'].mean(), inplace=True)
df['Solar.R'].fillna(df['Solar.R'].mean(), inplace=True)
df['Wind'].fillna(df['Wind'].mean(), inplace=True)
```

```python
df['Humidity'] = df['Humidity'].fillna(df['Humidity'].mode()[0])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
df["Humidity"] = lb.fit_transform(df["Humidity"])
df
```

```python
x = df.iloc[:, [0, 3]].values
y = df['Humidity']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```
